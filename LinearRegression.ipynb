```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error as rmse
Auto = pd.read_csv('Auto.csv').dropna()
Auto.head(5)
```

```python
Auto_shuffle = Auto.sample(frac=1).reset_index(drop=True)
n_rows = Auto_shuffle.shape[0]
training_rows = int(n_rows * .7)
train = Auto_shuffle[:training_rows]
test = Auto_shuffle[training_rows:]
print(train.shape, test.shape)
```

#Stats for the models on various values of X

```python
y = Auto['mpg'].to_numpy()
x_d = Auto['displacement'].to_numpy().reshape(-1, 1)
lr_d = LinearRegression().fit(x_d, y)
print("==========displacement============")
print("R^2 for displacement = ", lr_d.score(x_d, y))
print("intercept = ", lr_d.intercept_)
print("coefficients = ", lr_d.coef_)
y_hat_lr_d = lr_d.predict(x_d)
print("RMSE = ", rmse(y, y_hat_lr_d))
print()
print("============weight============")
x_w = Auto['weight'].to_numpy().reshape(-1, 1)
lr_w = LinearRegression().fit(x_w, y)
print("R^2 for weight = ", lr_w.score(x_w, y))
print("intercept = ", lr_w.intercept_)
print("coefficients = ", lr_w.coef_)
y_hat_lr_w = lr_w.predict(x_w)
print("RMSE = ", rmse(y, y_hat_lr_w))
print()
print("===========horsepower============")
x_h = Auto['horsepower'].to_numpy().reshape(-1, 1)
lr_h = LinearRegression().fit(x_h, y)
print("R^2 for horsepower = ", lr_h.score(x_h, y))
print("intercept = ", lr_h.intercept_)
print("coefficients = ", lr_h.coef_)
y_hat_lr_h = lr_h.predict(x_h)
print("RMSE = ", rmse(y, y_hat_lr_h))
print()
```

#Plotting the Data

```python
x_d1d = Auto['displacement'].to_numpy()
x_w1d = Auto['weight'].to_numpy()
x_h1d = Auto['horsepower'].to_numpy()

plt.figure()
plt.plot(x_d1d, y, "o", color="Red")
plt.xlabel("Displacement")
plt.ylabel("MPG")
plt.title("Displacement vs MPG")
plt.show()

plt.figure()
plt.plot(x_w1d, y, "x", color="Green")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Weight vs MPG")
plt.show()

plt.figure()
plt.plot(x_h1d, y, "o", color="Blue")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.title("Horsepower vs MPG")
plt.show()
```

```python
X = Auto[['displacement', 'weight', 'horsepower']].to_numpy()
mlr = LinearRegression().fit(X, y)
print("R^2 = ", mlr.score(X, y))
print("intercept = ", mlr.intercept_)
print("coefficients = ", mlr.coef_)
y_hat_mlr = mlr.predict(X)
print("RMSE = ", rmse(y, y_hat_mlr))
```
